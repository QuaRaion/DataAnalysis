## Цель работы

Для начала работы подключается библиотека `Pandas` и с помощью нее загружается датасет из `csv` файла, с указанием в качестве разделительного знака ",".
Выводятся первые 20 строк датасета с помощью метода `head(20)`.

```python
import pandas as pd
drivers = pd.read_csv('datasets/drivers2.csv', sep=',')
drivers.head(20)
```

С помощью метода `info()` оцениваются данные: есть ли пропуски, сколько всего строк, какие типы данных у столбцов.

```python
drivers.info()
```

Из вывода можно понять, что столбец `purpose` имеет пустые строки.

Применяется метод `describe()`, для оценки числовых столбцов.

По выводу команды можно определить:
*   `count`: количество заполненных строк;
*   `mean`: среднее значение;
*   `std`: стандартное отклонение;
*   `min`: минимальное число;
*   `25%`: 25-й процентиль;
*   `50%`: 50-й процентиль;
*   `75%`: 75-й процентиль;
*   `max`: максимальное число.

```python
drivers.describe()
```

В столбце `speed` среднее и максимальное значения равны `inf`, что означает, что они представляют из себя бесконечно большие числа. В дальнейшем необходимо убрать данные значения.

Выводятся названия столбцов с помощью `columns`.

```python
drivers.columns
```

Все заголовки приводятся к "верблюжьему" регистру при помощи метода `rename()`.

Для проверки названия столбцов выводятся.

```python
drivers = drivers.rename(columns={
    'START_DATE*':'startDate',
    'END_DATE*':'endDate',
    'CATEGORY*':'category',
    'START*':'start',
    'STOP*':'stop',
    'MILES*':'miles',
    'PURPOSE*':'purpose'
})

drivers.columns
```

Делается подсчет кол-ва пропусков в каждом столбце.

```python
drivers.isna().sum()
```

Пропуски имеются только в столбце `purposeRoute`. Так как в случае анализа оборота заказов в такси и других целей будут необходимы данные строки, будет нерационально удалять их, вместо этого пустые строки заполняются текстом "Other", так как цель поездки несущественно влиет на анализ осталльных показателей.

Для проверки выводятся первые 10 строк датасета и кол-во пропусков в столбцах.

```python
drivers['purpose'] = drivers['purpose'].fillna('Other')
drivers.head(10)
```

```python
drivers.isna().sum()
```

В столбце `speed` имеются 3 значения равные бесконечности ("inf"), а также есть значения, которые противоречат физическим способностям автомобилей, поэтому все экстремальные значения скорости (выбросы), которые больше, чем 130 миль/час (209 км/ч) заменяются на средние. Для этого необходимо посчитать среднее значение без значений "inf".

Нельзя удалять данные строки, чтобы можно было не терять информацию о кол-ве поездок по типам и категориям.

После этих действий `describe` корректно выведет информацию о столбце `speed`.

```python
import numpy as np

mean_speed = drivers['speed'][np.isfinite(drivers['speed'])].mean()

drivers.loc[drivers['speed'] > 130, 'speed'] = mean_speed

drivers.describe()
```

Производится проверка на наличие явных дубликатов и выводится их количество.

```python
drivers.duplicated().sum()
```

Дублирующиеся строки отсутствуют.


Производится проверка данных на наличие неявных дубликатов, для этого используется метод `unique()` для столбцов `category` и `purpose`, которые имеют категориальные переменные.

Столбцы `startTime`, `endTime`, `miles`, `time`, `speed`, `price` могут повторяться, а `start` и `stop` по условию можно не проверять на неявные дубликаты, поэтому данные столбцы не обрабатываются.


```python
print(drivers['category'].unique())
print(drivers['purpose'].unique())
```

Повторяющихся элементов в обоих столбцах не наблюдаются.

Вывод информации о датасете.

```python
drivers.info()
```

По выводу метода `info()` можно понять, что столбцы `startDate` и `endDate` не соответствуют  своему типу данных. Они записаны в формате срок (`object`), а должны быть в формате даты (`datetime`). Это исправляется с помощью метода `to_datetime()`.

Проверка правильности выполненных действий производится сразу.

```python
drivers['startDate'] = pd.to_datetime(drivers['startDate'], format='%Y-%m-%d %H:%M:%S')
drivers['endDate'] = pd.to_datetime(drivers['endDate'], format='%Y-%m-%d %H:%M:%S')

drivers.describe()
```

Предварительная обработка данных закончена.

### 2. Построение матрицы диаграмм рассеяния

Далее строится точечная диаграмма (матрица диаграмм рассеяния). Выполняется анализ полученной диаграммы и определяется показывает ли она в среднем определенную зависимость между переменными. Изучаются параметры и описываются взаимосвязи. Параметров не слишком много – нет потребности в создании нескольких графиков.

Подключается библиотека `seaborn`

```python
import seaborn as sns
sns.pairplot(drivers[['miles','category','time','speed', 'price']], hue="category")
```

Данная диаграмма отражает попарные взаимосвязи величин.

Выводы по диаграммам:
- в cвязи между милями и временем видна положительная корреляция, т.е. чем больше миль, тем больше затрачено времени. Это логично, так как длительность поездки, скорее всего, увеличивается с увеличением расстояния.
- в связи скорости и миль видно, что большинство поездок происходит на низкой скорости и на небольшие расстояния.
- в связи между скоростью и временем большинство значений сгруппированы около небольших значений скорости и времени, что указывает на короткие поездки с низкой скоростью;
- между ценой и временем поездки прослеживается положительная линейная связь, корреляция которой близка к единице, что показывает на сильную зависимость цены поездки от времени;
- в связи цены и миль также заметная положительная корреляционная свзяь, но с меньшей зависимотью переменных друг от друга, что указывает на не сильное влияние пройденнх миль на цену;
- между скоростью и ценой поездки связь практически не прослеживается.


### Оценка коэффициента корреляции и ковариации

Исследуются взаимосвязи между переменными с помощью оценки коэффициента корреляции и ковариации. Выполняется интерпретация результатов корреляции и ковариации для определения взаимосвязи между переменными.

```python
drivers.corr(numeric_only=True)
```

```python
drivers.cov(numeric_only=True)
```

Из вывода функций можно сделать следующие выводы:
- корреляция между `miles` и `time` сильная положительная. Это подтверждает наблюдение, что увеличение расстояния в милях связано с увеличением времени в пути. Чем дальше человек едет, тем больше времени он проводит в пути.
- между `price` и `miles` тоже очень высокая положительная корреляция. Это говорит о том, что более длительные поездки имеют более высокую цену.
- между `speed` и `time` практически нулевая корреляция. Это говорит о том, что скорость почти не влияет на время поездки.
- между между `speed` и `price` очень низкая положительная корреляция, что также указывает на то, что скорость не влияет на цену поездки.
- между `speed` и `miles` cлабая положительная корреляция, что значит, что связь между скоростью и расстоянием не сильная и сокрее всего на нее влияют много факторов.




Далее строится heatmap (тепловая карта корреляции).

```python
sns.heatmap(drivers.corr(numeric_only = True),
            annot = True, vmin=-1, vmax=1, center= 0, cmap = 'Blues')
```

С помощью диаграммы удобно прослеживать зависимости между всеми числовыми переменными.


### Построение графиков по заданию

Далее строятся графики по заданию.



**Задание 1.** Использовать seaborn. По группировке - `category` и количество поездок каждого типа (по цели маршрута - `purpose`) отфильтровать данные группировки по количеству поездок больше 2 и построить диаграмму.

```python
import matplotlib.pyplot as plt

dr2 = drivers.groupby(by=['purpose', 'category']).size().reset_index(name='count')
dr3 = dr2[dr2['count'] > 2]

plt.figure(figsize=(14, 5))
sns.set_style("whitegrid")
sns.barplot(x="purpose", y="count", hue="category", data=dr3)
plt.title("Диаграмма количества поездок по целям и категориям")
plt.xlabel("Цель поездки")
plt.ylabel("Количество поездок")
legend = plt.legend(loc='upper right')
legend.set_title("Категория")
plt.show()
```

Выводы по диаграмме:
- больше всего поездок по категории "Business" имеют цель "Others";
- больше всего поездок по категории "Person" имеют также цель "Others";
- по категории "Person" поездки имеют 2 цели - "Moving" и "Other";
- большинство поездок относится к категории "Business", что говорит о доминировании бизнес поездок;
- цель поездки "Other" имеет самое большое кол-во поездок, особенно в деловой категории, где их количество более чем в два раза превышает другие цели.
- "Person" поездки менее популярны, чем "Business"
- в категории "Person" выделяются только цели "Other" и "Meal/Entertain"

Таким образом, "Business" поездки значительно преобладают над "Person", и основная цель этих поездок — "Other".

**Задание 2** Использовать pandas и plot. По сводной таблице (pivot_table) - отобразить среднее количество пройденных миль по каждой цели поездки (PURPOSE). Оставить только маркеры в виде ★ зеленого цвета размеров 18.

```python
pv = drivers.pivot_table(index=['purpose'], values='miles', aggfunc='mean', dropna=False)

ax = pv.plot(
    style='*',
    markersize=18,
    color='green',
    grid=True,
    xlabel='Цель поездки',
    ylabel='Кол-во миль',
    title='График среднего числа пройденных миль по каждой цели поездки',
    figsize=(12, 3)
)

plt.xticks(rotation=90)
ax.set_xticks(range(len(pv.index)))
ax.set_xticklabels(pv.index)

plt.show()
```

На основе графика среднего числа пройденных миль по каждой цели поездки можно сделать вывод, что наибольшее количество миль связано с поездками по цели "Commute", что вероятно указывает на то, что поездки на большие расстояния чаще всего заказывают клиенты, место работы которых находится на большом расстоянии. Маршруты таких поездок примерно в 7 раз длиннее остальных.

**Задание 3** Использовать matplotlib. Построить круговую диаграмму, которая отображает процент по каждой цели поездки. Уберите из диаграммы количество поездок меньше 5.

```python
dr2 = drivers['purpose'].value_counts()
dr2 = dr2[dr2 > 5]

plt.figure(figsize=(8, 8))
plt.pie(dr2, labels = dr2.index, autopct='%1.1f%%', startangle=90, textprops={'size': '8'})
plt.legend(dr2.index, fontsize=10, bbox_to_anchor = (1.3, 1), loc='upper right')
plt.title("Распределение целей поездки", fontsize= 16)
plt.show()
```

По круговой диаграмме можно сделать следующие выводы:
- наибольшую долю занимает категория "Other", составившая 45.7% от общего числа поездок. Это может значить то, что у многих поездок не указывается цель, либо цель не относится к числу общих, которые есть.

1. Построить box-plot пройденных миль по цели поездки.

```python
purpose = drivers['purpose'].value_counts()[drivers['purpose'].value_counts() > 5].index
filtered_drivers = drivers[
    drivers['purpose'].isin(purpose) & (drivers['miles'] < 100)
]

plt.figure(figsize=(12, 10))
sns.boxplot(
    x = 'purpose',
    y = 'miles',
    data = filtered_drivers,
    hue = 'category',
    palette='Set3')

plt.xlabel('Цель поездки')
plt.ylabel('Мили')
plt.show()
```

По диаграмме можно сделать следующие выводы:
- по категории "Person" были совершены поездки только по цели "Other", что говорит либо о том, что при личных поездках пользователи не указывают цель, либо о том, что цели таких поездок не входят в список базовых целей;
- в каждой категории наблюдаются выбросы в значении пройденных миль, что может свидетельствовать о наличии редких и очень длинных поездок;

3. Построить первый график с помощью `matplotlib`. По группировке - `category` и количество поездок каждого типа (по цели маршрута - `purpose`) отфильтровать данные группировки по количеству поездок больше 2 и построить диаграмму.

Так как в `matplotlib` нет функции "hue" строятся отдельные столбцы для каждой категории с различными цветами. с помощью цикла "for".

```python
dr2 = drivers.groupby(by=['purpose', 'category']).size().reset_index(name='count')
dr3 = dr2[dr2['count'] > 2]
categories = dr3['category'].unique()
plt.figure(figsize=(12, 6))

for category in categories:
    subset = dr3[dr3['category'] == category]
    plt.bar(subset['purpose'], subset['count'], label=category)

plt.title("Диаграмма количества поездок по целям и категориям")
plt.xlabel("Цель поездки")
plt.ylabel("Количество поездок")
legend = plt.legend(loc='upper right')
legend.set_title("Категория")
plt.show()
```

По диаграмме можно сделать те же выводы, что и по первой, так как они одинаковые.
